```python
import pandas as pd
```

```python
df = pd.read_csv('flights.csv')
```

```python
df.info()
```

```python
# Check if there is a relation between airline and delay time
```

```python
airline_delays = df.groupby('name')['arr_delay'].mean()
```

```python
print(airline_delays.sort_values(ascending=False))
```

```python
#Here we can assume that it is more probable that when we fly with Frontier Airlines it's more probable that we will have a delay than with Alaska Airlines.
```

```python
#Now lets check if air time impacts arrival delay
```

```python
air_time_delay = df['air_time'].corr(df['arr_delay'])
print(f"Correlation coefficient: {air_time_delay:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['air_time'], df['arr_delay'], alpha=0.2, s=10)
plt.xlabel('Air Time (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.title('Scatter Plot: Air Time vs. Arrival Delay')
plt.show()
```

```python
#We can see there is almost no corelation
```
